```python
from bs4 import BeautifulSoup
import requests
import time
from random import randint

def scrape_sports_summaries(s):
    time.sleep(randint(0, 2))  # relax and don't let them know
    r = requests.get("https://www.stuff.co.nz/"+s)
    print(r.status_code)  # Print the status code
    content = r.text
    news_summaries = []
    soup = BeautifulSoup(content, "html.parser")
    st_divs = soup.findAll("div", {"class": "big-headline-top-stories section"})

    # print(st_divs)
    for st_div in st_divs:
        news_summaries.append(st_div.text)
    return news_summaries
```

```python
l = scrape_sports_summaries("sport")
for n in l:
    print(n)
```
